## ३. द्विघात समीकरण (Quadratic Equations)

ब्रह्मगुप्त ने ऋण मूलों को भी स्वीकार किया:

$ax^2 + bx + c = 0$ का हल:

$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

ऋण विविक्तकर (discriminant) → कोई वास्तविक मूल नहीं।

```python
सर्ग द्विघात_हल(a, b, c):
    """
    द्विघात समीकरण ax² + bx + c = 0 का हल।
    """
    यदि a == 0:
        # रैखिक समीकरण
        यदि b == 0:
            सर्गफल {'त्रुटि': 'कोई चर नहीं'}
        सर्गफल {
            'प्रकार': 'रैखिक',
            'मूल': [-c / b],
            'समीकरण': f'{b}x + {c} = 0'
        }
    
    विविक्तकर = b * b - 4 * a * c
    
    यदि विविक्तकर > 0:
        x1 = (-b + math.sqrt(विविक्तकर)) / (2 * a)
        x2 = (-b - math.sqrt(विविक्तकर)) / (2 * a)
        प्रकार = 'दो भिन्न वास्तविक मूल'
        मूल = [x1, x2]
    अनथा_यदि विविक्तकर == 0:
        x = -b / (2 * a)
        प्रकार = 'दो समान वास्तविक मूल'
        मूल = [x, x]
    अन्यथा:
        वास्तविक = -b / (2 * a)
        काल्पनिक = math.sqrt(-विविक्तकर) / (2 * a)
        प्रकार = 'काल्पनिक मूल'
        मूल = [f'{वास्तविक:.4f} + {काल्पनिक:.4f}i', f'{वास्तविक:.4f} - {काल्पनिक:.4f}i']
    
    सर्गफल {
        'समीकरण': f'{a}x² + {b}x + {c} = 0',
        'विविक्तकर': विविक्तकर,
        'प्रकार': प्रकार,
        'मूल': मूल
    }
```

```python
दर्शय("द्विघात समीकरण के हल:")
दर्शय("")

उदाहरण = [
    (1, -5, 6),    # x² - 5x + 6 = 0 → x = 2, 3
    (1, 2, -15),   # x² + 2x - 15 = 0 → x = 3, -5 (ऋण मूल!)
    (1, -4, 4),    # x² - 4x + 4 = 0 → x = 2, 2
    (1, 2, 5),     # x² + 2x + 5 = 0 → काल्पनिक
]

क्रमशः (a, b, c) अंतर्गत उदाहरण:
    फल = द्विघात_हल(a, b, c)
    दर्शय(f"{फल['समीकरण']}")
    दर्शय(f"  विविक्तकर = {फल['विविक्तकर']}")
    दर्शय(f"  {फल['प्रकार']}: {फल['मूल']}")
    दर्शय("")
```

```python
# द्विघात समीकरणों का दृश्य
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

उदाहरण = [
    (1, -5, 6, 'दो धन मूल'),
    (1, 2, -15, 'एक धन, एक ऋण मूल'),
    (1, -4, 4, 'दो समान मूल'),
    (1, 2, 5, 'काल्पनिक मूल'),
]

x = np.linspace(-8, 8, 200)

क्रमशः idx, (a, b, c, शीर्षक) अंतर्गत enumerate(उदाहरण):
    ax = axes[idx // 2][idx % 2]
    y = a * x**2 + b * x + c
    
    ax.plot(x, y, 'b-', linewidth=2)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    
    # मूल
    फल = द्विघात_हल(a, b, c)
    यदि 'काल्पनिक' not in फल['प्रकार']:
        क्रमशः मूल अंतर्गत फल['मूल']:
            ax.plot(मूल, 0, 'ro', markersize=10)
            ax.annotate(f'x={मूल:.2f}', (मूल, 0), textcoords='offset points', 
                       xytext=(5, 10), fontsize=10)
    
    ax.set_ylim(-20, 20)
    ax.set_title(f'{a}x² + {b}x + {c} = 0\n{शीर्षक}', fontsize=12)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## ४. ब्रह्मगुप्त की सर्वसमिका (Brahmagupta-Fibonacci Identity)

$$\boxed{(a^2 + nb^2)(c^2 + nd^2) = (ac \pm nbd)^2 + n(ad \mp bc)^2}$$

यह सर्वसमिका बताती है:
- दो "$a^2 + nb^2$" रूप की संख्याओं का गुणनफल भी उसी रूप का होता है!
- n = 1 के लिए: दो "वर्गों के योग" का गुणनफल भी "वर्गों का योग" है।

```python
सर्ग ब्रह्मगुप्त_सर्वसमिका(a, b, c, d, n=1):
    """
    (a² + nb²)(c² + nd²) = (ac ± nbd)² + n(ad ∓ bc)²
    """
    बायां = (a*a + n*b*b) * (c*c + n*d*d)
    
    # + version
    दायां_1 = (a*c + n*b*d)**2 + n*(a*d - b*c)**2
    
    # - version
    दायां_2 = (a*c - n*b*d)**2 + n*(a*d + b*c)**2
    
    सर्गफल {
        'बायां': बायां,
        'दायां_1': दायां_1,
        'दायां_2': दायां_2,
        'सत्यापन_1': बायां == दायां_1,
        'सत्यापन_2': बायां == दायां_2,
        'विघटन_1': f'({a*c + n*b*d})² + {n}×({a*d - b*c})²',
        'विघटन_2': f'({a*c - n*b*d})² + {n}×({a*d + b*c})²'
    }
```

```python
दर्शय("ब्रह्मगुप्त-फिबोनाची सर्वसमिका:")
दर्शय("")

# n = 1: दो वर्गों का योग
दर्शय("n = 1 (दो वर्गों का योग):")
a, b, c, d = 2, 3, 4, 1
फल = ब्रह्मगुप्त_सर्वसमिका(a, b, c, d, 1)

दर्शय(f"  ({a}² + {b}²)({c}² + {d}²) = {a*a + b*b} × {c*c + d*d} = {फल['बायां']}")
दर्शय(f"  = {फल['विघटन_1']} = {फल['दायां_1']} ✓")
दर्शय(f"  = {फल['विघटन_2']} = {फल['दायां_2']} ✓")

दर्शय("")

# n = -1: Pell equation connection
दर्शय("n = -1 (पेल समीकरण से सम्बन्ध):")
a, b, c, d = 3, 2, 3, 2
फल = ब्रह्मगुप्त_सर्वसमिका(a, b, c, d, -1)
दर्शय(f"  ({a}² - {b}²)({c}² - {d}²) = {a*a - b*b} × {c*c - d*d} = {फल['बायां']}")
दर्शय(f"  = {फल['विघटन_1']} = {फल['दायां_1']} ✓")
```

## ५. ब्रह्मगुप्त का द्वितीय क्रम अन्तर्वेशन (Second-Order Interpolation)

ब्रह्मगुप्त ने **Newton-Stirling** अन्तर्वेशन सूत्र का एक रूप दिया:

$$f(a + xh) \approx f(a) + \frac{x}{2}(D_1 + D_0) + \frac{x^2}{2}(D_1 - D_0)$$

जहाँ $D_0, D_1$ प्रथम अन्तर हैं।

```python
सर्ग ब्रह्मगुप्त_अन्तर्वेशन(y0, y1, y2, x):
    """
    ब्रह्मगुप्त का द्वितीय क्रम अन्तर्वेशन।
    
    y0, y1, y2 = x=0, 1, 2 पर ज्ञात मान
    x = अन्तर्वेशन बिंदु (0 < x < 2)
    """
    # प्रथम अन्तर
    D0 = y1 - y0
    D1 = y2 - y1
    
    # ब्रह्मगुप्त सूत्र
    फल = y1 + (x - 1) / 2 * (D0 + D1) + ((x - 1) ** 2) / 2 * (D1 - D0)
    
    सर्गफल {
        'मान': फल,
        'D0': D0,
        'D1': D1,
        'x': x
    }
```

```python
दर्शय("ब्रह्मगुप्त अन्तर्वेशन उदाहरण:")
दर्शय("")

# sin(30°) = 0.5, sin(45°) = 0.707, sin(60°) = 0.866
# sin(37.5°) ज्ञात करें (x = 0.5)

y0 = math.sin(math.radians(30))   # 0.5000
y1 = math.sin(math.radians(45))   # 0.7071
y2 = math.sin(math.radians(60))   # 0.8660

फल = ब्रह्मगुप्त_अन्तर्वेशन(y0, y1, y2, 0.5)

दर्शय(f"ज्ञात मान:")
दर्शय(f"  sin(30°) = {y0:.4f}")
दर्शय(f"  sin(45°) = {y1:.4f}")
दर्शय(f"  sin(60°) = {y2:.4f}")
दर्शय("")

दर्शय(f"sin(37.5°) का अनुमान:")
दर्शय(f"  ब्रह्मगुप्त: {फल['मान']:.6f}")
दर्शय(f"  सही मान:   {math.sin(math.radians(37.5)):.6f}")
दर्शय(f"  त्रुटि:    {abs(फल['मान'] - math.sin(math.radians(37.5))):.6f}")
```
